```python
import numpy as np
import matplotlib.pyplot as plt
#建议从tensorflow中引用
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.optimizers import SGD
```

```python
#生成两百个在-0.5到+0.5之间的随机数，并设置为列向量形式
x_data=np.linspace(-0.5,0.5,200)
x_data=x_data.reshape(200,1)
noise=np.random.normal(0,0.01,x_data.shape)
#y_data的值是x_data的平方加上noise
y_data=np.square(x_data)+noise
plt.scatter(x_data,y_data)
plt.show()
```

```python
model=Sequential()
#需要加上非线性的激活函数（tanh或ReLU）才能模拟非线性的数据，输入层一个神经元，隐藏层3个神经元，输出层1个神经元
#1-3-1
model.add(Dense(units=3,input_dim=1,activation='tanh'))
model.add(Dense(units=1))
#增加随机梯度下降法的学习率，可以减少迭代的次数
sgd=SGD(learning_rate=0.3)
model.compile(optimizer=sgd,loss='mse')
model.fit(x=x_data,y=y_data,epochs=3001,batch_size=100)
y_pred=model.predict(x_data)
plt.scatter(x_data,y_data)
plt.plot(x_data,y_pred,'r-')
plt.show()
```
